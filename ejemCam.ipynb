```python
import numpy as np
import cv2 as cv
import os
import time

rostro = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')
cap = cv.VideoCapture(0)

# Crear la carpeta para guardar las fotos
if not os.path.exists('fotos'):
    os.makedirs('fotos')

while True:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    rostros = rostro.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in rostros:
        frame = cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 5)
        frame = cv.rectangle(frame, (x+30, y+30), (x+w-30, y+h-100), (0, 255, 0), 5)
        w1 = int(w/2)
        h1 = int(h/2)

        # Capturar y guardar la foto solo cuando se detecta un rostro
        tiempo_actual = time.strftime("%Y%m%d_%H%M%S")
        foto_nombre = f'fotos/foto_{tiempo_actual}.png'
        cv.imwrite(foto_nombre, frame)
        print(f'Foto guardada: {foto_nombre}')

    cv.imshow('rostros', frame)

    k = cv.waitKey(1)
    if k == 27:
        break

cap.release()
cv.destroyAllWindows()

```

```python
import numpy as np
import cv2 as cv

rostro_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    rostros = rostro_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in rostros:
        frame = cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 5)

        # Coordenadas para reconocer la nariz
        nuevo_x_nariz = x + int(0.35 * w)
        nuevo_y_nariz = y + int(0.45 * h)
        ancho_nariz = int(0.3 * w)
        alto_nariz = int(0.2 * h)
        frame = cv.rectangle(frame, (nuevo_x_nariz, nuevo_y_nariz), (nuevo_x_nariz + ancho_nariz, nuevo_y_nariz + alto_nariz), (0, 0, 255), 5)

    cv.imshow('rostros', frame)

    k = cv.waitKey(1)
    if k == 27:
        break

cap.release()
cv.destroyAllWindows()

```
